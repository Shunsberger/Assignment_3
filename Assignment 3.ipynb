# Part 3

Groups the data in iris.txt by the species column


```python
# assigning species to data.groupby('Species') will make it easier to call this partitioned data set
species = data.groupby('Species')
```

# Part 3-1

Prints the summary statistics for each species in iris.txt

```python
species.describe()
```

Using species.describe() to print the summary statistics, allows a higher level summary of each species type. Based off of the data, it appears that PetalLength and PetalWidth are pretty unique for each species, and have a standard deviation that suggests there is not too much crossover in the data set. You will most likely be able to measure those two factors and be able to determine what species the flower is. If you look at SepalWidth and SepalLength, they do provide unique mean values for each species, but given the standard deviation, it is more likely that there the values have crossover for each species. It would be more difficult to identify the flower species using these measures. 

# Part 3-2

Generates a boxplot for each species in iris.txt

```python
# Uses original dataFrame, but groups by species to show a better comparison to other species
data.boxplot(by = 'Species', return_type = 'axes')

# Uses sorted dataFrame, and shows the variable measurements for the species in a unique boxplot per species
species.boxplot(return_type = 'axes')
```

I included two different ways to group the boxplot by species. I prefer the first version because it allows you to more easily compare the values of each species against the other two per variable measurement in each box. It helps confirm that PetalLength and PetalWidth have the least amount of overlap between species. 

The second boxplot includes each variable measurement in a unique boxplot per species, which makes it harder to compare between the species.
